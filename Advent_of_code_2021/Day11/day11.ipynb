```python
from collections import deque
with open("day11_input.txt", "r") as input_file:
    file_lines = input_file.readlines()
    G = []
    for line in file_lines:
        G.append([int(x) for x in list(line.strip())])

with open("test.txt", "r") as i:
    fl = i.readlines()
    T = []
    for line in fl:
        T.append([int(x) for x in list(line.strip())])

def day11(data):
    r = len(data)
    c = len(data[0])
    dr = [0,1,1,1,0,-1,-1,-1]
    dc = [-1,-1,0,1,1,1,0,-1]
    flashes = 0
    
    for step in range(500):
        step_flashes = 0
        #step 1
        for y in range(r):
            for x in range(c):
                data[y][x] +=1
        #step 2
        flashers = deque()
        for y in range(r):
            for x in range(c):
                if data[y][x] == 10:
                    flashers.append((y,x))
        while flashers:
            (y,x) = flashers.popleft()
            for d in range(8):
                rr = y+dr[d]
                cc = x+dc[d]
                if 0<=rr<r and 0<=cc<c and data[rr][cc] != 10:
                    data[rr][cc] +=1
                    if data[rr][cc] == 10:
                        flashers.append((rr,cc))

        #step 3
        for y in range(r):
            for x in range(c):
                if data[y][x] == 10:
                    flashes +=1
                    data[y][x] -= 10
                    step_flashes +=1
        step +=1
        print(f'step {step}: {flashes} total flashes')
        if step_flashes == 100:
            print(f'step {step} has all flashes!')
            break
```

```python
day11(G)
```
