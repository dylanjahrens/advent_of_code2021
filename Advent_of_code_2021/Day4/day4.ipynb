```python
#Parsing data
with open("day4_input.txt", "r") as input_file:
    file_lines = input_file.readlines()
    input_list = [line.strip() for line in file_lines]
    input_list.append('')

grid_list = []
index = 0
for i, string in enumerate(input_list):
    if string == '':
        grid_list.append(input_list[index:i])
        index = i+1

bingo_numbers = grid_list[0][0]
bingo_balls = bingo_numbers.split(',')
bingo_cards = grid_list[1:]
bingo_grid = [] # each card is a list with 5 lists for each line, each of 5 numbers is a string 
#grid-card-line-number

for card in bingo_cards:
    card_grid = []
    for line in card:
        line = line.split(' ')
        for char in line:
            if len(char) < 1:
                line.remove(char)
        card_grid.append(line)
    bingo_grid.append(card_grid)
    card_grid = []
    

def day4_1():

    for ball in bingo_balls:
        for card in bingo_grid:
            for line in card:
                for i, number in enumerate(line):
                    if number == ball:
                        line.remove(number)
                        line.insert(i, '#')
                        
                #across win check
                if line == ['#', '#', '#', '#', '#']:
                    winning_board = card
                    winning_ball = ball
                    board_sum = 0
                    for line in winning_board:
                        for num in line:
                            if num != '#':
                                board_sum += int(num)
                    return board_sum * int(winning_ball)
                
            #down win check
            i = 0
            while i < 5:
                if card[0][i] == card[1][i] == card[2][i] == card[3][i] == card[4][i] == '#':
                    winning_board = card
                    winning_ball = ball
                    board_sum = 0
                    for line in winning_board:
                        for num in line:
                            if num != '#':
                                board_sum += int(num)
                    return winning_board, winning_ball, board_sum * int(winning_ball)
                                
                else:
                    i+=1

                    
                    
```
